Com estes dados, podemos construir gráficos de linha com a evolução diária dos Time Controls mais jogados, _Blitz_ e _Bullet_

```python
def find_ratings(r_ , time_control):
    for item in r_:
        if item['name'] == time_control:
            ratings = item['points']
            break
    return ratings
        
blitz_ratings = find_ratings(r_, 'Blitz')
bullet_ratings = find_ratings(r_, 'Bullet')        
```

Definimos uma função que devolve os ratings para cada tipo de time control e agora vamos visualizar a evolução

```python
#Convert time list to datetime
dates = []
ratings = []
for rating in blitz_ratings:
    dates.append(datetime.datetime(rating[0], rating[1]+1, rating[2]))
    ratings.append(rating[3])

# Settings for first subplot (Blitz)    
plt.subplot(2, 1, 1)
plt.plot(dates,ratings, 'ro-')
plt.title('Blitz and Bullet Ratings')
plt.ylabel('Blitz')
plt.gcf().autofmt_xdate()

#Convert time list to datetime
dates = []
ratings = []
for rating in bullet_ratings:
    dates.append(datetime.datetime(rating[0], rating[1]+1, rating[2]))
    ratings.append(rating[3])

# Settings for second subplot (Bullet)     
plt.subplot(2, 1, 2)
plt.plot(dates,ratings, 'go-')
plt.xlabel('Date')
plt.ylabel('Bullet')
plt.gcf().autofmt_xdate()
```

Para conseguirmos identificar quais as aberturas que têm mais sucesso precisamos de perceber qual a taxa de sucesso que o utilizador tem com essas aberturas.

```python
#Initialization of lists with the opening and the outcome
openings = []
outcomes = []

# This piece of code iterates over the list of games, 
# on the game moves, it takes the first 3 moves by each player, 
# on the game description, it takes the result, from the user perspective (win or other)
for item in range(len(r.split('\n\n'))-1):
    if item % 2 != 0:
        openings.append(r.split('\n\n')[item].split(' 4.')[0])
    else:
        if user in r.split('\n\n')[item].split('Black')[0]:
            if (r.split('\n\n')[item].split('Result')[1].split(']')[0].strip()) == '"1-0"':
                outcomes.append(1)
            else:
                outcomes.append(0)
        else:
            if (r.split('\n\n')[item].split('Result')[1].split(']')[0].strip()) == '"1-0"':
                outcomes.append(0)
            else:
                outcomes.append(1)            

# Once we've got our data as we want it, we zip it into a list with the opening and outcome associated
openings_outcomes = list(zip(openings, outcome))
```

We will now start by scrapping each opening FEN. FEN is the standard representation of a chess position and is easily recognized as the default computer notation.

First we will scrape the page, and then get the part that includes the FEN position.

```python
#Function that returns the html from the get request
def get_data(url, parse):
    """
    returns a parsed html 
    """
    raw_html = requests.get(url, parse).content
    return BeautifulSoup(raw_html)

#Function that gets the FEN position
def get_fen(url):
    raw_html = requests.get(url).content
    html = BeautifulSoup(raw_html)
    fen = html.find('input', {"name":'fen_position'})['value']
    return fen

# Now, we will apply the get_fen function and add a column with the FEN to our dataframe
df_scrapping['FEN'] = df_scrapping['link'].map(lambda x : get_fen(x))

df_scrapping
```

Já temos o FEN. Agora, vamos procurar pelos restantes dados que pretendemos.

Para avaliarmos a qualidade de cada jogada, é sempre importante percebermos qual a avaliação do computador. 

Contudo, muitas vezes, aquilo que o computador avalia não é tão percetível para os jogadores, os chamados _computer moves_. 

Nesse sentido, é também um fator, percebermos quantos jogos já foram jogados naquela posição e quais os resultados obtidos.

Estes resultados, têm sempre que ser ponderados tendo em conta um número razoável de observações, portanto, também precisamos do número de jogos.


Para termos estes dados, vamos iterar por cada link do nosso dataset e vamos recolher os respetivos dados.

```python
#Function to get HTML
def get_html(url):
    raw_html = requests.get(url).content
    html = BeautifulSoup(raw_html)
    tables = html.find_all('table', {'cellpadding':'0', 'width':"100%",'cellspacing':'0'})[1].contents
    return html, tables

#Function to get the Current evaluation
def get_evaluation(html):
    evaluation = html.find('div', attrs={'style': 'font-style:italic;font-size:12px;width:240px;text-align:center;'})
    evaluation = evaluation.text.replace('Stored Analysis:', '').strip()
    return evaluation

#Function to get the Suggested Moves
def get_move_suggested_from_tag(tag):
    if tag.find('a') != None:
        return tag.find('a').text.replace('3... ','')

    
#Function to get the number of games played for given position
def get_game_number_from_tag(tag):
    if tag.find('td',  attrs={'style': 'padding-right:3px;'}) != None:
        return int(tag.find('td',  attrs={'style': 'padding-right:3px;'}).text.replace('\xa0','').replace(',',''))

#Function to get the rates     
def get_result_rate(tag, background):
    result = tag.find('td', attrs={'align':'center', 'background':background})
    if  result != None:
        if '%' not in result.text:                
            return 'Rare'
        else:
            return result.text.strip()

#Function to get the engine evaluation
def get_engine_eval(tag):
    if tag.find('td', attrs={'style':"font-style:italic;font-size:12px;"}) != None:
        return tag.find('td', attrs={'style':"font-style:italic;font-size:12px;"}).text.strip()
```

```python
#Initialize Lists
games = []
current_evaluations = []
evaluations = []
moves = []
white_rates = []
draw_rates = []
black_rates = []


#Background style win rate percentage for each possible result
background_white = "/images/opbkw.gif.pagespeed.ce.Orrko29Ywq.gif"
background_draw = "/images/opbkd.gif.pagespeed.ce.eddg-sJv3Z.gif"
background_black = "/images/opbkb.gif.pagespeed.ce.lw33PDRZL2.gif"


#Iterate page to scrape the interesting information
for link in df_scrapping['link']:
    game_number = []
    move_suggested = []
    white_win_rate = []
    draw_rate = []
    black_win_rate = []
    engine_evaluation = []    
    
    html, tables = get_html(link)
    
    current_evaluations.append(get_evaluation(html))
    
    
    for s in tables:
        if isinstance(s, Tag):
            if len(move_suggested) < 3 and get_move_suggested_from_tag(s) != None:
                move_suggested.append(get_move_suggested_from_tag(s))
            if len(game_number) < 3 and get_game_number_from_tag(s) != None:
                game_number.append(get_game_number_from_tag(s))
            if len(white_win_rate) < 3 and get_result_rate(s, background_white) != None:
                white_win_rate.append(get_result_rate(s, background_white))
            if len(draw_rate) < 3 and get_result_rate(s, background_draw) != None:
                draw_rate.append(get_result_rate(s, background_draw))
            if len(black_win_rate) < 3 and get_result_rate(s, background_black) != None:
                black_win_rate.append(get_result_rate(s, background_black))
            if len(engine_evaluation) < 3 and get_engine_eval(s) != None:    
                engine_evaluation.append(get_engine_eval(s))
                
    #Save info to List            
    games.append(game_number)
    evaluations.append(engine_evaluation)
    moves.append(move_suggested)
    white_rates.append(white_win_rate)
    draw_rates.append(draw_rate)
    black_rates.append(black_win_rate)

    
    
```

```python
#Add columns from our lists to the datatable
df_scrapping['Engine Evaluation'] = current_evaluations
df_scrapping['Top 3 Moves'] = moves
df_scrapping['Played Games Top 3 Moves'] = games
df_scrapping['Evaluations Top 3 Moves'] = evaluations
df_scrapping['White Win Rate Top 3 Moves'] = white_rates
df_scrapping['Draws Top 3 Moves'] = draw_rates
df_scrapping['Black Win Rate Top 3 Moves'] = black_rates

pd.reset_option('colheader_justify')
df_scrapping[df_scrapping.columns[5:12]]
```
